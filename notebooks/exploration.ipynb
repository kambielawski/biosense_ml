# Data Exploration

Use this notebook to explore the raw dataset, inspect samples, and prototype transforms.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# Update this path to your local dataset location
RAW_DATA_DIR = Path.home() / "BIOSENSE Dropbox" / "BIOSENSE Team Folder" / "biosense_communication_interface" / "prod" / "archive"
```

```python
# List files in the dataset
if RAW_DATA_DIR.exists():
    files = sorted(RAW_DATA_DIR.rglob("*"))[:20]
    for f in files:
        print(f"{f.suffix:>6}  {f.stat().st_size / 1024:.1f} KB  {f.relative_to(RAW_DATA_DIR)}")
else:
    print(f"Directory not found: {RAW_DATA_DIR}")
```

```python
# Load and display a sample image
image_extensions = {".jpg", ".jpeg", ".png", ".tif", ".tiff"}
image_files = [f for f in RAW_DATA_DIR.rglob("*") if f.suffix.lower() in image_extensions]
print(f"Found {len(image_files)} images")

if image_files:
    img = Image.open(image_files[0])
    print(f"Size: {img.size}, Mode: {img.mode}")
    plt.figure(figsize=(8, 8))
    plt.imshow(np.array(img))
    plt.title(image_files[0].name)
    plt.axis("off")
    plt.show()
```
